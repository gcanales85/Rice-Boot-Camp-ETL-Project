```python
import pandas as pd
import requests
import json
import os
from pprint import pprint
import pymongo
```

```python
csv_file = os.path.join("Resources", "Opioid_Prescribing_Rate_webscrape.csv")
customer_data_df = pd.read_csv(csv_file)
customer_data_df.head()
```

```python
prescription_df = customer_data_df[['State', 'State ABBR', 'Prescribing Rate','Year']].copy()
prescription_df.head()
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.opioidabuse_db
collection = db.opioidabuse_prescription
collection.drop()
```

```python
for index, rows in prescription_df.iterrows():
    dictionary = {
    'state': rows [0],
    'state abbr': rows [1],
    'prescribing rate': rows [2],
    'year': str(rows [3])
        
    }
    
    collection.insert_one(dictionary)
```

```python
for result in collection.find():
    pprint(result)
```
