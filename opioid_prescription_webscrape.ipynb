```python
# dependencies
import pandas as pd

#web scraping the needed years
url = "https://www.cdc.gov/drugoverdose/maps/rxstate2006.html"
table2006 = pd.read_html(url)
df2006 = table2006[0]
df2006["Year"]="2006"
df2006=df2006.rename(columns = {'2006 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2007.html"
table2007 = pd.read_html(url)
df2007 = table2007[0]
df2007["Year"]="2007"
df2007=df2007.rename(columns = {'2007 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2008.html"
table2008 = pd.read_html(url)
df2008 = table2008[0]
df2008["Year"]="2008"
df2008=df2008.rename(columns = {'2008 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2009.html"
table2009 = pd.read_html(url)
df2009 = table2009[0]
df2009["Year"]="2009"
df2009=df2009.rename(columns = {'2009 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2010.html"
table2010 = pd.read_html(url)
df2010 = table2010[0]
df2010["Year"]="2010"
df2010=df2010.rename(columns = {'2010 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2011.html"
table2011 = pd.read_html(url)
df2011 = table2011[0]
df2011["Year"]="2011"
df2011=df2011.rename(columns = {'2011 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2012.html"
table2012 = pd.read_html(url)
df2012 = table2012[0]
df2012["Year"]="2012"
df2012=df2012.rename(columns = {'2012 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2013.html"
table2013 = pd.read_html(url)
df2013 = table2013[0]
df2013["Year"]="2013"
df2013=df2013.rename(columns = {'2013 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2014.html"
table2014 = pd.read_html(url)
df2014 = table2014[0]
df2014["Year"]="2014"
df2014=df2014.rename(columns = {'2014 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2015.html"
table2015 = pd.read_html(url)
df2015 = table2015[0]
df2015["Year"]="2015"
df2015=df2015.rename(columns = {'2015 Prescribing Rate':'Prescribing Rate'})

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2016.html"
table2016 = pd.read_html(url)
df2016 = table2016[0]
df2016["Year"]="2016"
df2016=df2016.rename(columns = {'2016 Prescribing Rate':'Prescribing Rate'})


# years 2017 and 2018 have differently structured columns, will need to fix before final combination
url = "https://www.cdc.gov/drugoverdose/maps/rxstate2017.html"
table2017 = pd.read_html(url)
df2017 = table2017[0]
df2017["Year"]="2017"

url = "https://www.cdc.gov/drugoverdose/maps/rxstate2018.html"
table2018 = pd.read_html(url)
df2018 = table2018[0]
df2018["Year"]="2018"
```

```python
#fixing year 2017's headers by importing into another dataframe
df2017fixed= pd.DataFrame()
df2017fixed["State"] = df2017["Unnamed: 0_level_0","State"]
df2017fixed["State ABBR"] = df2017["Unnamed: 1_level_0","Abbreviation"]
df2017fixed["Prescribing Rate"] = df2017["Opioid Prescribing Rate per 100","2017"]
df2017fixed["Year"] = df2017["Year"]
#df2017fixed.head()
```

```python
#fixing year 2018's headers by importing into another dataframe
df2018fixed= pd.DataFrame()
df2018fixed["State"] = df2018["Unnamed: 0_level_0","State"]
df2018fixed["State ABBR"] = df2018["Unnamed: 1_level_0","Abbreviation"]
df2018fixed["Prescribing Rate"] = df2018["Opioid Prescribing Rate per 100","2018"]
df2018fixed["Year"] = df2018["Year"]
#df2018fixed.head()
```

```python
# combining all the dataframes, exporting to csv to be consumed later.
bin = [df2006,df2007,df2008,df2009,df2010,df2011,df2012,df2013,df2014,df2015,df2016,df2017fixed,df2018fixed]
df_allyears =pd.concat(bin, ignore_index=True)
df_allyears.to_csv("Resources/Opioid_Prescribing_Rate_webscrape.csv")
```
